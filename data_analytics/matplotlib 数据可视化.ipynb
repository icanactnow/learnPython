## 1.1 简单绘制

## 1.2 设置图形的属性

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0, 5, 0, 20])# 设置横纵坐标最小最大值[xmin, xmax, ymin, ymax]
plt.title("it's a plot chart")# 设置标题
plt.show()
```

```python
# 再一张图中绘制多条曲线
import math
import numpy as np
t = np.arange(0, 2.5, 0.1) # 生成又0到2.5间隔为0.1的数
y1 = map(math.sin, math.pi*t)# 高阶函数map作用是将函数作用到每一个元素上
y2 = map(math.sin, math.pi*t+math.pi/2)
y3 = map(math.sin, math.pi*t-math.pi/2)
# plt.plot(t, list(y1), 'b*', t, list(y2), 'g^', t, list(y3), 'ys')# 绘制顺序是（横坐标，纵坐标，表示符号，横坐标，纵坐标，表示符号，。。。）
# plt.show()
plt.plot(t, list(y1), 'b--', t, list(y2), 'g', t, list(y3), 'y-.')
plt.show()
```
